# calculate the corresponding vapor pressure

```python
epsilon = 0.622
e = lev * q / (q + epsilon)
print(e.shape)
```

# Calculate the ambient dewpoint given the vapor pressure.

```python
Td = (243.5 * np.log(e/6.112)) / (17.67 - np.log(e/6.112)) + 273.15
print(Td[0,:].shape)
```

```python
def getParcelT(lev,q,t):
    epsilon = 0.622
    e = lev * q / (q + epsilon)
    Td = (243.5 * np.log(e/6.112)) / (17.67 - np.log(e/6.112)) + 273.15

    lev1 = lev.tolist() * units.hPa
    t1 = t.tolist() * units.kelvin
    Td1 = Td.tolist() * units.kelvin
    p,tt,tdd,tp = calc.parcel_profile_with_lcl(lev1, t1, Td1)
    return p,tt,tdd,tp
```
